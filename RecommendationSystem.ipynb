```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('Datasets/tmdb_5000_movies.csv')
credits = pd.read_csv('Datasets/tmdb_5000_credits.csv')
```

```python
movies = movies.merge(credits, on='title') # merging both the datasets (on basis of 'title')
```

```python
movies.head(1)
```

```python
movies.info()
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']] # relevant cols
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.shape
```

```python
movies.drop_duplicates(inplace=True)
movies.shape
```

```python
movies.iloc[0].genres
```

```python
import ast # will help convert string of lists (given) to a list 
```

```python
def convert_to_simple_list(x):
    l = []
    for i in ast.literal_eval(x):
        l.append(i['name'])
    return l
```

```python
movies['genres'] = movies['genres'].apply(convert_to_simple_list)
```

```python
movies.head(1)
```

```python
movies['keywords'] = movies['keywords'].apply(convert_to_simple_list)
```

```python
movies.head(1)
```

```python
def convert_to_simple_list_keywordsCOL(x):
    l = []
    counter = 0
    for i in ast.literal_eval(x):
        if counter != 3:
            l.append(i['name'])
            counter += 1
        else: 
            break
    return l
```

```python
movies['cast'] = movies['cast'].apply(convert_to_simple_list_keywordsCOL)
```

```python
movies.head(1)
```

```python
movies['crew'][0]
```

```python
def fetch_director(x):
    l = []
    for i in ast.literal_eval(x):
        if i['job'] == 'Director':
            l.append(i['name'])
            break
    return l
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head(1)
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
movies.head(1)
```

```python
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
movies.head(1)
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head(1)
```

```python
new_df = movies[['movie_id','title','tags']]
```

```python
new_df
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df.head(3)
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```

```python
new_df.head(3)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names_out() # most frequent words existing
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    x = []
    for i in text.split():
        x.append(ps.stem(i))

    return " ".join(x)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
similarity
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:11]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
